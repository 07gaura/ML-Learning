```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

**Line Graph**

---



```python
x= [1,2,3,4]
y=[2,4,6,8]
x2 = np.arange(1,5,0.5)
plt.figure(figsize=(5,3),dpi=300)
plt.title('Our First Graph',fontdict={'fontsize':20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks(x)
plt.yticks(y)



#plt.plot(x,y,label='2x',color='red',linewidth=2,linestyle='--',marker='.',markersize=10,markeredgecolor='blue')
#short hand notation
#'colormarkerline'
plt.plot(x,y,'b^--',label='2x')
plt.plot(x2,x2**2,'r.-')
plt.legend()
plt.show()
```

**Bar Chart**

---



```python
labels = ['A','B','C']
values = [1,4,3]

bars = plt.bar(labels,values)

patterns=['/','o','*']
for bar in bars:
  bar.set_hatch(patterns.pop(0))
#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
plt.figure(figsize=(6,4))

plt.show()
```

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
gas.head()
```

```python
filt = (gas['Year']>=1990) & (gas['Year']<=2000)
data = gas[filt]
plt.title('Gas Prices between 1990 and 2000 in usd')
for country in data:
  if country!='Year':
    plt.plot(data.Year,data[country],'-',label=country)
#plt.plot(data.Year,data.Canada,'g.-',label="canada")
plt.xlabel('Year')
plt.ylabel('Price')
plt.legend(loc='upper left')
plt.show()
```

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head(5)
```

**Histograms**

---



```python
bins = np.arange(50,110,10)
plt.hist(fifa.Overall,bins=bins)
plt.xticks(bins)
plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of player skills in fifa')
plt.show()
```

```python
fifa.columns
```

```python
left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]

labels = ['left','right']
colors=['Green','Red']
plt.pie([left,right],labels=labels,colors=colors,autopct='%.2f %%')
plt.title('Foot preference of FIFA Players')
plt.show()
```

```python
obj = fifa.dtypes['Weight']
obj
```

```python
li = []
for x in fifa.Weight:
  if type(x)==str:
    li.append(int(x.strip('lbs')))
  else:
    li.append(x)
fifa['Weight']=li
```

```python
obj = fifa.dtypes['Weight']
obj
```

```python
light = fifa.loc[fifa.Weight<125].count()[0]
light_medium = fifa.loc[(fifa.Weight<125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight<150) & (fifa.Weight < 175)].count()[0]
medium_heavy=fifa.loc[(fifa.Weight<175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight<200].count()[0]

weights=[light,light_medium,medium,medium_heavy,heavy]
labels = ['Under 25','125-50','150-75','175-200','Over 200']
explode = (.1,.1,.1,.1,.1)
plt.pie(weights,labels=labels,autopct="% .2f %%", pctdistance=0.8,explode=explode)
plt.show()
```

```python
barcelona = fifa.loc[fifa.Club=='FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club=='Real Madrid']['Overall']

labels = ['FC Barcelona','Madrid']
plt.boxplot([barcelona,madrid],labels=labels)
plt.title('Professional Soccer Team Comparison')
plt.ylabel('Fifa Overall Rating')
plt.show()
```

```python
import pandas as pd
import re
```

```python
raw_data = ""
with open("WhatsApp Chat with BD QD.txt","r") as file:
  raw_data = file.read()
```

```python
print(raw_data)
```

```python
data = re.split('\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s\w+\s-\s',raw_data)[2:]
```

```python
data
```

```python
users = []
messages = []

for i in data:
  x = i.split(":")
  if len(x)==1:
    continue
  users.append(x[0])
  messages.append(x[1])
```

```python
date = re.findall('\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s\w+',raw_data)[2:]
```

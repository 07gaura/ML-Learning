```python
people = {
    "first":["gaurav","kunal", "pranay"],
    "last": ["lokhande","pawar","pawar"],
    "email": ["gourav@gmail.com","kunal@gmail.com","pranay@gmail.com"]
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df.columns = [x.upper() for x in df.columns]
df
```

```python
df.columns = [x.lower() for x in df.columns]
df
```

```python
df.rename(columns={'first':'first_name','last':'last_name'},inplace=True)
```

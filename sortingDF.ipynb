```python
people = {
    'first': ['gaurav','pranay','kunal','akash'],
    'last': ['lokhande','pawar','pawar','sahu'],
    'email':['gaurav@gmail.com','pranay@gmail.com','kunal@gmail.com','akash@gmail.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df.sort_values(by='last',ascending=False)
```

```python
df.sort_values(by=['last','first'],ascending = False)
```

```python
df.sort_values(by=['last','first'],ascending = [False,True],inplace=True)
```
